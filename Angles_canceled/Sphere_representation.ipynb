```python
import numpy as np
import matplotlib.pyplot as plt
import qutip as qt
%matplotlib qt5
```

```python
theta = np.random.uniform()*np.pi
phi = np.random.uniform()*2*np.pi
t1 = theta
p1 = phi

t2 = theta
p2 = np.pi-phi

psi1 = [np.sin(t1)*np.cos(p1), np.sin(t1)*np.sin(p1), np.cos(t1)]
psi2 = [np.sin(t2)*np.cos(p2), np.sin(t2)*np.sin(p2), np.cos(t2)]


fig1 = plt.figure()

b1 = qt.Bloch(fig=fig1)
b1.add_vectors(psi1)
b1.add_vectors(psi2)
b1.zlabel = [r'$z$', '']


b1.add_points([0.5*np.cos(np.linspace(0,p1,100)), 0.5*np.sin(np.linspace(0,p1,100)),np.zeros(100)])
b1.add_points([0.5*np.sin(np.linspace(0,t1,100)), np.zeros(100), 0.5*np.cos(np.linspace(0,t1,100))])
#b1.render()

b1.show()
#b1.fig.savefig('Bloch.png', bbox_inches="tight")
```

```python
fig2 = plt.figure()
b2 = qt.Bloch(fig=fig2)
b2.add_vectors(psi1)
b2.add_vectors(psi2)
b2.view = [0,0]
b2.zlabel = [r'$z$', '']
b2.show()
#b2.fig.savefig('Bloch_y.png', bbox_inches="tight")
```
